**Grubbs Test** <br>
Uji Grubbs mendeteksi satu outlier pada satu waktu (nilai tertinggi atau terendah), sehingga hipotesis nol dan alternatifnya adalah sebagai berikut: <br>
H0: Nilai tertinggi bukan outlier <br>
H1: Nilai tertinggi adalah outlier <br>
jika kita ingin menguji nilai tertinggi, atau: <br> 
H0: Nilai terendah bukanlah outlier <br>
H1: Nilai terendah adalah outlier <br>
jika kita ingin menguji nilai terendah. <br>
<br>
Jika nilai p kurang dari 0.05 maka hipotesis nol ditolak dan kami akan menyimpulkan bahwa nilai terendah/tertinggi adalah outlier . Sebaliknya, jika nilai p lebih besar atau sama dengan tingkat signifikansi, maka hipotesis nol tidak ditolak, dan kita akan menyimpulkan bahwa berdasarkan data, kita tidak menolak hipotesis bahwa nilai terendah/tertinggi tidak sebuah outlier.

```r
test2 <- grubbs.test(Dataset$Product_id)
test2
test1 <- grubbs.test(Dataset$Product_id, opposite = TRUE)
test1
```

####**Coefficient of Determination (R-Squared)**



R-squared adalah angka yang berkisar antara 0 sampai 1 yang mengindikasikan besarnya kombinasi variabel independen secara bersama – sama mempengaruhi nilai variabel dependen. Secara umum, semakin besar nilai R-squared atau semakin mendekati angka 1 pada suatu model regresi, semakin baik variabel penjelas mampu memprediksi nilai variabel respon.

####**Mean Absolute Error (MAE)**

Mean Absolute Error (MAE) merepresentasikan perbedaan antara nilai asli dan nilai prediksi yang diekstraksi dengan rata-rata perbedaan absolut pada kumpulan data. Semakin kecil nilai MAE menunjukkan kecocokan model yang lebih baik. Ini akan lebih berguna ketika kita membandingkan dua model yang berbeda.

###**Analisis dan Diskusi**

####**1. Linear Regression**


Regresi linear yang kami gunakan adalah regresi linear berganda dikarenakan banyaknya variabel. Pengujian regresi linear menggunakan uji asumsi klasik, yang antara lain terdiri dari: uji normalitas, uji heteroskedastisitas, uji auto korelasi, dan uji multikolinearitas. Regresi linier digunakan untuk menyelesaikan masalah regresi. Regresi linier memperkirakan variabel dependen ketika ada perubahan pada variabel independen. Dalam regresi linier, nilai outputnya kontinu.

####**2. Logistic Regression**

Regresi logistik dapat digunakan untuk mengklasifikasikan dua kelas. Ini juga dikenal sebagai klasifikasi biner. Regresi logistik adalah model statistik yang memprediksi probabilitas hasil yang hanya dapat memiliki dua nilai. Dalam uji regresi logistik, uji prasyarat atau kualitas data yang dilakukan antara lain: uji keseluruhan model (overall model fit), uji kelayakan model menggunakan Hosmer and Lemeshow Test. Regresi linier memperkirakan variabel dependen ketika ada perubahan pada variabel independen. Dalam regresi logistik, nilai outputnya bersifat diskrit.

####**3. Poisson Regression**

Regresi Poisson termasuk dalam regresi nonlinier yang variabel responnya dimodelkan sebagai distribusi Poisson. Variabel respon dalam regresi Poisson berasal dari data cacahan yang kejadiannya jarang terjadi. Dalam model regresi Poisson terdapat asumsi yang harus dipenuhi yakni equidispersi yang berarti nilai variansi dari variabel respon Y yang diberikan oleh X = x harus sama dengan nilai meannya. Regresi poisson menggunakan metode maksimum likelihood dalam menduga parameternya. Pada regresi poisson, responnya menyatakan jumlah suatu kejadian.

####**4. Perbandingan antara Regresi Linear dengan Regresi Logistik**

Dari kedua Regresi tersebut dapat kita ketahui bahwa regresi linear dan regresi logistik memiliki kesamaan pada setiap Coefficients Intercept-nya menghasilkan hasil yang sama pada kedua regresi tersebut.

####**5. Perbandingan antara Regresi Linear dengan Regresi Poisson**

Dari kedua regresi diatas yaitu regresi Linear dan regresi Poisson dapat disimpulkan bahwa tidak adanya kesamaan antara kedua regresi tersebut, yaitu regresi linear dan regresi poisson. Perbedaan kedua regresi tersebut dapat terlihat jelas dari bagian Coefficients-nya pada bagian Intercept memiliki hasil yang tidak berbeda jauh diantara keduanya. Pada regresi Linear mendapatkan hasil estimasi 6.909 x 102 dan pada regresi Poisson mendapatkan hasil 6.512x 102 sehingga memiliki selisih perbedaan 0.397 x 102. Regresi poisson ini menggunakan metode maksimum likelihood untuk menduga atau mengestimasi parameternya.
